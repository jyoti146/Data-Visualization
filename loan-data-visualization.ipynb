```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('loan.csv')

#replacing blank by NaN
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values='NaN',strategy='mean',axis=0)

#finding missing values in Credit_History
a = list(df['Credit_History']).count(1)
b = list(df['Credit_History']).count(0)
if a>b:
    df['Credit_History'].replace(np.NaN,1,inplace=True)
else:
    df['Credit_History'].replace(np.NaN,0,inplace=True)

#finding missing values in Self_Employed
x = list(df['Self_Employed']).count('Yes')
y = list(df['Self_Employed']).count('No')
if x>y:
    df['Self_Employed'].replace(np.NaN,'Yes',inplace=True)
else:
    df['Self_Employed'].replace(np.NaN,'No',inplace=True)

#finding missing values in Loan_Amount_Term
c = df['Loan_Amount_Term'].mean()
df['Loan_Amount_Term'].replace(np.NaN,c,inplace=True)

#finding missing values in LoanAmount
d = df['LoanAmount'].mean()
df['LoanAmount'].replace(np.NaN,d,inplace=True)

#finding missing values in Gender
p = list(df['Gender']).count('Male')
q = list(df['Gender']).count('Female')
if p>q:
    df['Gender'].replace(np.NaN,'Male',inplace=True)
else:
    df['Gender'].replace(np.NaN,'Female',inplace=True)
    
#part1:Number of applicants applied for loan
l = len(list(df['Loan_ID']))
print("Number of applicants applied for loan =",l,end='\n\n')

#part2:Gender wise count of Loan Applicants shown as pie chart
e = [p,q]
labels = 'Male','Female'
colors = ['crimson','gold']
explode = (0.1,0)
plt.title('Gender wise count of Loan Applicants')
plt.pie(e,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.show()

#part3:Property area wise count of Loan Applicants shown as bar chart
u = list(df['Property_Area']).count('Urban')
v = list(df['Property_Area']).count('Semiurban')
w = list(df['Property_Area']).count('Rural')
objects = ('Urban','Semiurban','Rural')
y_pos = np.arange(len(objects))
performance = [u,v,w]
plt.bar(y_pos,performance,color='green')
plt.xticks(y_pos,objects)
plt.ylabel('Count')
plt.xlabel('Property_Area')
plt.title('Property area wise count of loan applicants')
plt.show()

#part4:Show all Loan_ID who are not self employed
f = list(df['Loan_ID'])
g = list(df['Self_Employed'])
print("Loan_ID who are not self employed:",end='\n\n')
for i in range(len(g)):
    if g[i]=='No':
        print(f[i])
print(end='\n')

#part5:Show the Loan_ID and Applicant Income as bar chart
z = list(df['ApplicantIncome'])
j = list(df['Loan_ID'])
y = []
x = []
for i in range(len(z)):
    if z[i]>=2000 and z[i]<=2500:
        x.append(j[i])
        y.append(z[i])
y_pos = np.arange(len(x))
performance = y
plt.bar(y_pos,performance,color='crimson')
plt.ylabel('Salary')
plt.xlabel('Respective Loan_ID')
plt.title('Loan_ID and ApplicantIncome as bar chart')
plt.show()

#part6:List all the applicants who do not have any credit history
h = list(df['Credit_History'])
print("The applicants without credit history are:",end='\n\n')
for i in range(len(h)):
    if h[i]==0:
        print(f[i])
```
